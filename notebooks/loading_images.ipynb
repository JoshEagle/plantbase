# Reading the csv 

```python
# as usual, this filepath only works if you are running a notebook from plantbase/notebooks
plants_df = pd.read_csv("../plantbase/data/plants_with_image_dimensions_v1").drop(columns="Unnamed: 0")
```

```python
plants_df.head()
```

# Creating subfolders for each genus

```python
np.unique(plants_df.genus.values)
```

```python
def create_folders():
    """takes each unique genus as a category,
    then creates a new folder for that genus if one doesn't already exist"""
    categories = np.unique(plants_df.genus.values)
    for genus in categories:
        folder = r"../raw_data/train/" + str(genus)
        if not os.path.exists(folder):
            print('created folder', genus)
            os.mkdir(folder)
        else:
            print(str(genus), 'exists!')
```

```python
create_folders()
```

# Moving files into their correct genus

```python
def move_files():
    """
    iterates over the rows in plants_df,
    moves the original file into the new_filename location"""
    
    failed = 0
    for i, row in plants_df.iterrows():
        new_filename = rf"../raw_data/train/{row.genus}/{row.filename}"
        old_filename = rf"../raw_data/train/{row.filename}"
        if not os.path.exists(new_filename):
            try:
                os.rename(old_filename, new_filename)
                print(f"moved {row.filename} to {row.genus}")
            except:
                failed += 1
        else:
            print(f"{row.filename} is in {row.genus}")
            
    print(f"failed on {failed} files")
```

```python
move_files()
```

# Processing images

```python
from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
directory = r"../raw_data/train"
```

```python
tmp_dataset = image_dataset_from_directory(
    directory, labels='inferred', label_mode='int', class_names=None,
    color_mode='rgb', batch_size=32, image_size=(256, 256), shuffle=True, seed=123,
    validation_split=None, subset=None, interpolation='bilinear', follow_links=False
)
```

```python
tmp_dataset
```

```python
class_names = tmp_dataset.class_names
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in tmp_dataset.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```
