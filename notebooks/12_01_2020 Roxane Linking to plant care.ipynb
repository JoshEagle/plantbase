# Importing the plant care data

```python
import pandas as pd

data=pd.read_csv('../plantbase/scraping/plant_info.csv').drop(columns='Unnamed: 0')
```

```python
data.head()
```

```python
data['Genus_class'] = data['Genus'].str.split(' ').str[0]
data.head()
```

```python
class_numbers={'Ajuga': 0,
 'Allium': 1,
 'Campanula': 2,
 'Cirsium': 3,
 'Crataegus': 4,
 'Gentiana': 5,
 'Geranium': 6,
 'Iris': 7,
 'Malva': 8,
 'Narcissus': 9,
 'Ophrys': 10,
 'Rosa': 11,
 'Trifolium': 12,
 'Verbascum': 13,
 'Veronica': 14,
 'Viola': 15}
```

```python
data['class']=data['Genus_class'].map(class_numbers)
data
```

```python
data.set_index('class')
```

```python
data.to_csv(r'../plantbase/scraping/plant_info_updated.csv')
```
