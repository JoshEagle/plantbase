```python
## This can take 5-10minutes!

import os
import pandas as pd 
import xml.etree.ElementTree as et 

df_cols = ["id","family", "genus", "species", "content"]
rows = []

# I can't remember how to do absolute file paths so for now just put your own one in! 
directory = r'C:\Users\joshu\code\Joshymitzu\plantbase\raw_data\train'
for filename in os.listdir(directory):
    if filename.endswith(".xml"):
        #print(filename)
        xtree = et.parse(os.path.join(directory, filename))
        xroot = xtree.getroot() 
        s_id=xroot.find('MediaId').text
        s_family= xroot.find('Family').text
        s_genus= xroot.find('Genus').text
        s_species= xroot.find('Species').text
        s_content = xroot.find('Content').text
        rows.append({"id": s_id,"family": s_family, "genus": s_genus, "species": s_species, "content":s_content})
        print(f'parsed {s_genus}')
    else:
        continue
        
out_df = pd.DataFrame(rows, columns = df_cols)
```

# Cleaning the dataset

## Searching the dataset for genus names

## Creating list of popular plants to include

```python
# popular plants selected VERY scientifically
popular_plants=['Daffodils','Campanula','Verbascum','Crataegus','Geranium','Ophrys','Hyacinth','Hydrangea' ,'Iris' ,'Jasmine' ,'Lavender' ,'Rhododendron' ,'Roses' ,'Snowdrops' ,'Tulip' ,'Wisteria']
```

```python
# checking how many of our popular plants names are already included in the dataset

plant_incl=[]
plant_notincl=[]
for plant in popular_plants:
    if plant in out_df['genus'].values:
        plant_incl.append(plant)
    else: 
        plant_notincl.append(plant)
        
print (f" {len(plant_incl)} popular plants are already in the dataset, {len(plant_notincl)} are not.")
```

```python

# dictionary mapping common plant names to latin names used in dataset
plant_names= {'Daffodils':'Narcissus', 'Roses': 'Rosa', 'Tulip': 'Tulipa', 'Snowdrops': 'Galanthus', 'Jasmine': 'Jasminum', 'Lavender': 'Lavandula', 'Hyacinth': 'Hyacinthoides'}

#looping over popular plants list and replacing the English name with the latin name, 
# if the english name appears in the plant_names dictionary
for index, plant in enumerate(popular_plants):
    if plant in plant_names.keys():
        popular_plants[index]=plant_names[plant]
    else: 
        pass

len(popular_plants)
```

## Creating a filtered dataframe

```python
#including only genuses that are in our popular plants list
priority_genus=out_df[out_df.genus.isin(popular_plants)]
```
